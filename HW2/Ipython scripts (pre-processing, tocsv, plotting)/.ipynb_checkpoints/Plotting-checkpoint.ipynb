### NN

```python
with h5py.File('log_hyper_embedding.f5','r') as hf:
    print('List of arrays in this file: \n', hf.keys())
    loss = hf.get('loss_train')
    loss = np.array(loss)
    training_acc = hf.get('training_accuracy')
    training_acc = np.array(training_acc)
    train_time = hf.get('training_time')
    train_time = np.array(train_time)
    valid_acc = hf.get('valid_accuracy')
    valid_acc = np.array(valid_acc)
```

```python
hyp = [40,50,60,80,100,120]
loss_nn1 = [0.30321519416855, 0.16526912050898, 0.13882272715113, 0.12316241705001, 0.11365823466305, 0.10569034252408, 0.10009360907648, 0.094368546652581, 0.090007589763883, 0.085937003812082,  0.082638819089707,  0.080688000292496, 0.077611036617596, 0.074708496370951, 0.072265615799489, 0.070, 0.069, 0.0685,0.0683,0.068]
```

```python
plt.figure(figsize=(10,8))
for i in range(6):
    plt.plot(range(4,21,4),loss[i,:],label=('Hidden parameter = %s' % hyp[i]))
    plt.plot(range(4,21,4),loss[i,:],'bo',color='black',alpha=0.6)
#plt.plot(range(4,21),loss_lr[3:],label = 'LogReg')
plt.plot(range(4,21),loss_nn1[3:],label = 'NN without embeddings', color = 'black')
plt.plot(range(4,21,4),[loss_nn1[j-1] for j in range(4,21,4)],'bo',color='black',alpha=0.6)
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Negative LogLikelihood')
plt.xlim(3,21)
plt.xticks(range(4,21,4))
plt.title('Evolution of the Loss as a function of the number of Epochs for the Neural Network Model\n with word embeddings\n')
plt.show()
```

```python
from matplotlib import rc
rc('text', usetex=False)
```

```python
plt.figure(figsize=(10,8))
for i in range(6):
    plt.plot(range(4,21,4),training_acc[i,:],label=('Hidden parameter = %s' % hyp[i]))
    plt.plot(range(4,21,4),training_acc[i,:],'bo',color='black',alpha=0.6)
plt.legend(loc=4)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.xlim(3,21)
plt.xticks(range(4,21,4))
plt.title('Evolution of the Accuracy on Training data as a function of the number of Epochs for the Neural Network Model\n with word embeddings\n')
plt.show()
```

```python
plt.figure(figsize=(10,8))
for i in range(6):
    plt.plot(range(4,21,4),valid_acc[i,:],label=('Hidden parameter = %s' % hyp[i]))
    plt.plot(range(4,21,4),valid_acc[i,:],'bo',color='black',alpha=0.6)
plt.legend(loc=4)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.xlim(3,21)
plt.xticks(range(4,21,4))
plt.title('Evolution of the Accuracy on Validation data as a function of the number of Epochs for the Neural Network Model\n with word embeddings\n')
plt.show()
```
